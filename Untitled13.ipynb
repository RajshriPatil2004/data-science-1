Data Preprocessing


```python
import pandas as pd
df = pd.concat([train], axis=0)
train_len = len(train)
print(df.head())
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
#deop and delete a column

df= df.drop(columns=['Cabin'],axis=1)
```

```python
df['Age'].mean()
```

```python
#fill missing values using mean of column
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

```python
#drop a duplicate value and column
df.duplicated().sum()

# Remove duplicates
df.drop_duplicates(inplace=True)
```
